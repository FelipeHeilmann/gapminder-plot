```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('./data.csv')

data.columns = data.columns.str.strip()

x = data['2019']
y = data['1992']
size = data.loc[:, '1992':'2019'].mean(axis=1)

norm_size = (size - size.min()) / (size.max() - size.min()) * 2000

color = data.loc[:, '1992':'2019'].std(axis=1)

plt.figure(figsize=(12, 8))
scatter = plt.scatter(x, y, s=norm_size, c=color, cmap='viridis', alpha=0.6, edgecolors="w", linewidth=0.5)
plt.colorbar(scatter, label='Desvio padro da porcentagem de homems do setor de servio')

plt.title('Porcentagem de homems do setor de servio (1992 vs 2019)')
plt.xlabel('Porcentagem em 2019')
plt.ylabel('Porcentagem em 1992')

for i, country in enumerate(data['country']):
    plt.annotate(country, (x[i], y[i]), fontsize=9, alpha=0.75)

plt.show()
```
